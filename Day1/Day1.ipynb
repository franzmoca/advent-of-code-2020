```
def part2(list):
  length = len(list)
  for i in range(length):
      for j in range(length):
        for k in range(length):
            sum = list[i] + list[j] + list[k]
            if sum == 2020:
              print(i,j,k, list[i], list[j], list[k])
              print("Solution is :", list[i] * list[j] * list[k] )
              return;
part2(list)
          
```
